# Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from fbprophet import Prophet
```

# Importing the dataset 

```python
avocado_df = pd.read_csv('avocado.csv')
```

```python
avocado_df
```

# Exploring the data 

```python
avocado_df.head()
```

```python
avocado_df = avocado_df.sort_values("Date") #sorting according to 'Date'
```

```python
avocado_df
```

```python
plt.figure(figsize=(10,10))
plt.plot(avocado_df['Date'], avocado_df['AveragePrice'])    #plt.plot(x,y)
```

```python
# Bar Chart to indicate the number of regions 
plt.figure(figsize=[25,12])
sns.countplot(x = 'region', data = avocado_df)   #count elements based on regions in our dataset

plt.xticks(rotation = 45)  #rotates the x axis plots by 45 degrees  ----> look at X-axis
```

```python
# Bar Chart to indicate the year
plt.figure(figsize=[25,12])
sns.countplot(x = 'year', data = avocado_df)    #counting based on year on our datset
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]   #taking only the 'Date' and 'AveragePrice' from the dataset
```

```python
avocado_prophet_df
```

# Making Predictions 

### using facebook prophet `

```python
avocado_prophet_df = avocado_prophet_df.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
avocado_prophet_df
```

```python
#using Prophet() from fbprophet

model = Prophet()
model.fit(avocado_prophet_df)
```

```python
# Forcasting into the future
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)
```

```python
forecast
```

```python
figure = model.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure_trend = model.plot_components(forecast)
```

```python
avocado_df_sample
```
